```python
import pandas as pd
```

```python
shisuu = pd.read_csv('./shisuu_new.csv')
```

```python
s = shisuu.merge(shisuu, on='race_id')
```

```python
s = s[s['horse_race_id_x'] != s['horse_race_id_y']]
```

```python
s['sum_score'] = s['score_x'] + s['score_y']
s
```

```python
haitou = pd.read_csv('./csv_new2/race_detail.csv')
haitou = haitou.set_index('race_id')
```

```python
h = haitou[['1着馬番', '2着馬番', '馬連', '馬単']].merge(s, on='race_id')
```

```python
umaren = h[
    (
        (h['h_num_x'] == h['1着馬番'])
        &
        (h['h_num_y'] == h['2着馬番'])      
    )
    |
    (
        (h['h_num_x'] == h['2着馬番'])
        &
        (h['h_num_y'] == h['1着馬番'])      
    )
]
```

```python
# umaren = umaren.reset_index()

pd.set_option('display.max_rows', None)
au = umaren[['sum_score', '馬連']].groupby('sum_score').count()
au_haitou = umaren[['sum_score', '馬連']].groupby('sum_score').sum()
```

```python
au = au.merge(au_haitou, on='sum_score')
```

```python
au
```

```python
ah = h[['sum_score', '馬連']].groupby('sum_score').count()
```

```python
auh = ah.merge(au, how='left', on='sum_score').fillna(0)
```

```python
auh['的中率'] = auh['馬連_x'] / auh['馬連']
auh['回収率'] = (auh['馬連_y'] / (auh['馬連'] * 100)) * 100
auh[['回収率', '的中率']].sort_values('sum_score', ascending=False)
```

```python
import numpy as np

s = auh.reset_index()
s = s[(s['sum_score'] % 5) == 0]

auh = auh.reset_index()
auh['group_id'] = auh['sum_score'].map(lambda x: np.digitize(x, np.array(s['sum_score'])))
```

```python
auhg = auh[['group_id']].set_index('group_id')

auhg['点数'] = auh[['group_id', '馬連']].groupby('group_id').sum().fillna(0)
auhg['配当'] = auh[['group_id', '馬連_y']].groupby('group_id').sum()

auhg = auhg.drop_duplicates()
auhg['回収率'] = (auhg['配当'] / (auhg['点数'] * 100)) * 100
auh[['group_id', 'sum_score']].merge(auhg, on='group_id').sort_values('sum_score', ascending=False)
```
